Program 1:

Create the below pattern using nested for loop in Python.

    *
    * *
    * * *
    * * * *
    * * * * *
    * * * *
    * * *
    * *
    *

```python
#many ways to print above pattern
#using nested for loop  
for i in range(1,10):
    if i<6:
        for j in range(1,6):
            if j<=i:
                print('* ',end="")
    else:
        for k in range(i,10):
            print('* ',end="")
    print()
```

Write a Python program to reverse a word after accepting the input from the user.
