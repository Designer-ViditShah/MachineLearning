Program 1: 

Write a program which will find all such numbers which are divisible by 7 but are not a
multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed
in a comma-separated sequence on a single line.

```python
for i in range(2000,3201):
    if i%7==0 and i%5!=0:
        if i==3199:
            print(i)
            break
        print(i,end=",")
```

Program 2:

Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

```python
first_name = input("Please enter your first name: ")
last_name = input("Please enter your last name: ")
#if reversing means reversing the order of the input strings
print("Type one : \n",last_name,end=" , ")
print(first_name)
#if reversing means reversing the indices of the input strings along with its order
print("Type two : \n",last_name[::-1],end=" , ")
print(first_name[::-1])
```

```python
import math    #contains value of pi
diameter = 12   #given in centimeter
radius = diameter//2 #divides the number to integer value
volume = (4/3)*math.pi*radius*radius*radius
#round function is used to round the float value
#second argument is for number of decimal points to be included
print("Volume of the sphere with diameter 12 cm is : ",round(volume,2))
```
