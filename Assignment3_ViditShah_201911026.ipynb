PROBLEM 1.1:

Implementing myreduce() function 

```python
def add(a,b):
    return a+b
def mult(a,b):
    return a*b
def myreduce(fun1,lst):
        s2 = lst[0]
        for i in range(len(lst)-1):
            n = fun1(s2,lst[i+1])
            s2 = n
        return s2
myList = [1,2,3,4,5]
x = myreduce(add,myList)
print("Sum of list is : ",x)
x = myreduce(mult,myList)
print("Product of list is : ",x)
```

PROBLEM 1.2:

Implementing myfilter() function

```python
import math
def prime(a):
    if a==2 or a==3:
        return True
    elif a>3:
        for j in range(2,a):
            if a%j==0:
                return False
        return True
    else:
        return False
def myfilter(fun1,lst):
    s2 = []
    for i in lst:
        if fun1(i)==True:
            s2.append(i)
    return s2
myList = []
for k in range(1,100):
    myList.append(k)
x = myfilter(prime,myList)
print("Prime number between 1-100 are : ",x)
```
